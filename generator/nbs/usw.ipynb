```python
from pathlib import Path

async def visit_html_files(directory):
    base_path = Path(directory)
    for file_path in base_path.rglob("*"):
        #print(f"{file_path=}")
        if file_path.suffix.lower() == ".html":
            await process_html_file(file_path)

async def process_html_file(file_path):
    #print(f"Processing: {file_path}")
    # Add your code here to process each HTML file
    html = open(file_path, "r", encoding="iso-8859-1").read()
    title, text = html_to_text(html)
    if title is not None:
        title = title.string.strip()
    if not title:
        title = file_path.stem
    print(f"saving {file_path=} {title=}")
    try:
        email = await db.email.create({
            "name": title,
            "html": html,
            "text": text,
        })
    except Exception as e:
        print(f"{e=}")
```

```python
from bs4 import BeautifulSoup

def html_to_text(html):
    soup = BeautifulSoup(html, "html.parser")
    return soup.title, soup.get_text()
```
